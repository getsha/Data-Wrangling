## Calculating the mean,mode and median

```python
import pandas as pd
import statistics
import re

# Make a copy of the data
filtered_data_copy = filtered_data.copy()

# Calculate statistics for 'discounted_price'
discounted_price_mean = filtered_data_copy['discounted_price'].mean()
discounted_price_median = filtered_data_copy['discounted_price'].median()
discounted_price_mode = statistics.mode(filtered_data_copy['discounted_price'])
discounted_price_variance = filtered_data_copy['discounted_price'].var()

# Calculate statistics for 'actual_price'
actual_price_mean = filtered_data_copy['actual_price'].mean()
actual_price_median = filtered_data_copy['actual_price'].median()
actual_price_mode = statistics.mode(filtered_data_copy['actual_price'])
actual_price_variance = filtered_data_copy['actual_price'].var()

# Calculate statistics for 'discount_percentage'
discount_percentage_mean = filtered_data_copy['discount_percentage'].mean()
discount_percentage_median = filtered_data_copy['discount_percentage'].median()
discount_percentage_mode = statistics.mode(filtered_data_copy['discount_percentage'])
discount_percentage_variance = filtered_data_copy['discount_percentage'].var()

# Clean 'rating' column and convert it to numeric values
rating_cleaned = filtered_data_copy['rating'].apply(lambda x: re.sub('[^0-9]', '', x))
rating_numeric = pd.to_numeric(rating_cleaned)

# Calculate statistics for 'rating'
rating_mean = rating_numeric.mean() / 10
rating_median = rating_numeric.median() / 10
rating_mode = statistics.mode(rating_numeric) / 10
rating_variance = rating_numeric.var() / 10

# Create a DataFrame with the calculated statistics
data = {
    'Statistic': ['Mean', 'Median', 'Mode', 'Variance'],
    'Discounted Price': [discounted_price_mean, discounted_price_median, discounted_price_mode, discounted_price_variance],
    'Actual Price': [actual_price_mean, actual_price_median, actual_price_mode, actual_price_variance],
    'Discount Percentage': [discount_percentage_mean, discount_percentage_median, discount_percentage_mode, discount_percentage_variance],
    'Rating': [rating_mean, rating_median, rating_mode, rating_variance]
}
statistics_table = pd.DataFrame(data)

# Display the table
print(statistics_table)
```

## Linear Regression Model 

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

amazon_data = filtered_data.copy()

# This removes any data row with null data
amazon_data = amazon_data.dropna()

amazon_data.loc[:, 'discount_percentage'] = amazon_data.loc[:, 'discount_percentage'].replace('%', '', regex=True).astype(float)

X = amazon_data.drop(columns=['product_id', 'discounted_price', "actual_price", 'category', 'rating', 'rating_count', 'product_name', 'review_title', 'review_id', 'review_content', 'img_link', 'product_link', 'user_name', 'user_id', 'about_product'])
y = amazon_data['rating']

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)

# Creating and training the model
model = LinearRegression()
model.fit(X_train, y_train)

# Predicting on the test set
y_pred = model.predict(X_test)

# Compute mean squared error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# Sorting the values
sorted_indices = y_test.argsort()
y_test_sorted = y_test.iloc[sorted_indices]
y_pred_sorted = y_pred[sorted_indices]

# Plotting the scatter graph with sorted values
plt.scatter(y_test_sorted, y_pred_sorted)
plt.xlabel("Actual Rating")
plt.ylabel("Predicted Rating")
plt.title("Actual vs. Predicted Rating")
plt.show()
```

## KNN regression model

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

amazon_data = filtered_data.copy()

# This removes any data row with null data
amazon_data = amazon_data.dropna()

amazon_data.loc[:, 'discount_percentage'] = amazon_data.loc[:, 'discount_percentage'].replace('%', '').astype(float)

X = amazon_data.drop(columns=['product_id', 'discounted_price', 'actual_price', 'category', 'rating', 'rating_count', 'product_name', 'review_title', 'review_id', 'review_content', 'img_link', 'product_link', 'user_name', 'user_id', 'about_product'])

# Convert 'rating' column to float
y = amazon_data['rating'].astype(float)

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)

# Creating and training the model
model = KNeighborsRegressor(n_neighbors=5)  # Set the number of neighbors (k) to 5
model.fit(X_train, y_train)

# Predicting on the test set
y_pred = model.predict(X_test)

# Compute mean squared error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# Generate smooth trendline
X_range = np.arange(len(y_test))
trendline = model.predict(X_test)

# Plotting the line graph with smooth trendline
plt.plot(X_range, y_test, label="Actual Rating ")
plt.plot(X_range, y_pred, label="Predicted Rating")
plt.plot(X_range, trendline, label="Trendline", linestyle='--')
plt.xlabel("Percentage Discount")
plt.ylabel("Rating ")
plt.title("Actual vs. Predicted Rating with Trendline")
plt.legend()
plt.show()
```
